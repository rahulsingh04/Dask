```python
import numpy as np 
import dask.array as da
```

```python
size_tuple = (500,500)
chunks_tuple = (10, 500)
```

```python
da_arr = da.random.randint(10 , size = size_tuple, chunks =chunks_tuple)
da_arr2 = da.random.randint(10 , size = size_tuple, chunks =chunks_tuple)
```

```python
def random_func(x):
    return np.mean((((x*2).T)**2),axis = 0)
```

```python
gufoo = da.gufunc(random_func, signature = "(i)->()", 
                 output_dtypes = float, 
                 vectorize = True)
```

```python
random_op_arr = gufoo(da_arr)
random_op_arr.compute()
```

```python
random_op_arr.shape
```

### You Can Also Use as gufunc as decorator for gufunc 

```python
@da.as_gufunc(signature= "(m, n),(n,j)->(m,j)" , output_dtypes = int, allow_rechunk = True)
def random_func(x, y):
    return np.matmul(x,y)**2
```

```python
da_arr3 = da.random.randint(10 , size =(200, 100), chunks = (10, 100))
da_arr4 = da.random.randint(10 , size =(100, 300), chunks = (5, 5))
```

```python
random_matmul = random_func(da_arr3, da_arr4)
```

```python
random_matmul.compute()
```
